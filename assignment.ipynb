## Task 7 : Merging and Joining DataFrames

```python
# Merge two DataFrames
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']},
                    index=['K0', 'K1', 'K2', 'K3'])
df2 = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=['K0', 'K1', 'K2', 'K3'])
```

```python
df1
```

```python
df2
```

```python
joined_df = df1.join(df2)
```

```python
joined_df
```

## Task 8: Working with Dates and Times

```python
# Create a datetime index for the DataFrame
joined_df['Date'] = pd.date_range(start = '1/1/2024', periods=len(joined_df), freq='m')
```

```python
joined_df
```

```python
# Convert a column to datetime and extract date components
joined_df['Date'] = pd.to_datetime(joined_df['Date'])
```

```python
joined_df['Year'] = joined_df['Date'].dt.year
```

```python
joined_df['Month'] = joined_df['Date'].dt.month
```

```python
joined_df['Days'] = joined_df['Date'].dt.day
```
